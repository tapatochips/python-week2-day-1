### If Statements

```python
# Available operators: Greater(>), Less(<),Equal(==)
# Greater or Equal(>=), Less or Equal (<=)

# Truth Tree:
# T && F = F
# T && T = T
# T || F = T
# F || T = T
# F || F = F
num1 = 25
num2 = 15
if num1 == num2:
    print('this is true')
elif num1 > num2:
    print('num1 is bigger than num2')
else:
    print('equal')
```

##### Continue Statement

```python
# will continue to next iteration
```

```python
for number in range(10):
    if number == 5:
        print('skip')
        continue
    print(number)
print('done')
```

##### Break Statement

```python
# will break out of current loop
```

```python
for number in range(10):
    if number == 5:
        print('skip')
        break
    print(number)
print('done')
```

##### Pass Statement

```python
# mostly used as a placeholder, and will continue on same iteration
```

```python
for number in range(10):
    if number == 5:
        print('skip')
        pass
    print(number)
print('done')
```

##### Double For Loops

##### Looping 'While True'

```python
answer = 50
guess = input('guess a number between 1-100')
while True:
    if guess == answer:
        print('you got it')
        break
    else:
        print(f'you guessed {guess}')
        print('try again')
              
    
```

##### While & For Loops Used Together

```python
num = 0
while num < 5:
    print('\nWhile loop iteration', num)
    for i in range(2):
        print('\n for loop iteration')
    num +=1
```
